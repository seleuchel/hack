```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(1)
```

```python
X_xor = np.random.randn(200,2)
y_xor = np.logical_xor(X_xor[:,0] >0,
                      X_xor[:, 1] >0)
print(y_xor)
print()
```

```python
y_xor = np.where(y_xor, 1, -1)
print(y_xor)
```

```python
plt.scatter(X_xor[y_xor == 1, 0],
           X_xor[y_xor == 1,1],
           c='b', marker='x',
           label='1')

plt.scatter(X_xor[y_xor == -1, 0],
           X_xor[y_xor == -1, 1],
                c='r',
                marker='s',
                label='-1')

plt.tight_layout()
plt.show()
```

```python
# sdkfh
from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVC
svm = SVC(kernel='rbf',gamma=1.0 ,C=10.0, random_state=1)
# fit : 학습
svm.fit(X_xor, y_xor)

#??
plot_decision_regions(X_xor,
                     y_xor,
                     clf=svm)

plt.xlabel('petal length [atandarized]')
plt.ylabel('width')

plt.legend(loc='upper left')

plt.tight_layout()
plt.show()


# sdkfh
from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVC
svm = SVC(kernel='rbf',gamma=0.010 ,C=10.0, random_state=1)
# fit : 학습
svm.fit(X_xor, y_xor)

#??
plot_decision_regions(X_xor,
                     y_xor,
                     clf=svm)

plt.xlabel('petal length [atandarized]')
plt.ylabel('width')

plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```
